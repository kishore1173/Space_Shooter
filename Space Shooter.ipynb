```python
import pygame 
import random
import math
```

```python
pygame.init()
```

```python
pygame.font.init()
screen = pygame.display.set_mode((800,600))
pygame.display.set_caption("Space Fight") #flaticon

jetx = 370
jety = 480
jetx_change =0
jet = pygame.image.load("aircraft.png")
jet = pygame.transform.scale(jet,(50,50))


planex = []
planey = []
planey_change = []
plane = []





no_enemy = 1

def enemy(no_enemy):
    
    
    e = pygame.image.load("jet.png")
    e = pygame.transform.scale(e,(50,50))
    e = pygame.transform.rotate(e,180)
    for i in range(no_enemy):
        plane.append(e)
        planex.append(random.randint(50,750))
        planey.append(random.randint(0,100))
        planey_change.append(0)

enemy(2)
"""
planex = []
planey = []
planey_change = []
plane = []
e = pygame.image.load("jet.png")
e = pygame.transform.scale(e,(50,50))
e = pygame.transform.rotate(e,180)
for i in range(6):
    plane.append(e)
    planex.append(random.randint(50,750))
    planey.append(random.randint(0,100))
    planey_change.append(0)
"""


space = pygame.image.load("Space.png")
space = pygame.transform.scale(space,(800,600))


bulletx = 370
bullety = 460
bulletx_change = 0
bullety_change = 1
bullet = pygame.image.load("bullet1.png")
bullet = pygame.transform.scale(bullet,(20,20))
state = "ready"

score = 0
font = pygame.font.SysFont("timesnewroman.ttf",20)
tx = 10
ty = 10
def score_val(x,y):
    sc = font.render("Score:"+ str(score),True,(255,255,255))
    screen.blit(sc,(x,y))
def game_over():
    f1 = pygame.font.SysFont("timesnewroman.ttf",100)
    text = f1.render("GAME OVER",True,(255,0,0))
    screen.blit(text,(200,200))

def jetplayer(x,y):
    screen.blit(jet,(x,y))

def planeplayer(x,y,i):
     screen.blit(plane[i],(x,y))

def fired(x,y):
    global state
    state = "Fire"
    screen.blit(bullet,(x,y))
    
def collision(x1,y1,x2,y2):
    r = math.sqrt((math.pow(x2-x1,2)) + (math.pow(y2-y1,2)))
    if r < 35:
        return True
    else:
        return False


run = True
while run:
    screen.fill((0,0,0))
    screen.blit(space,(0,0))
    
    for event in pygame.event.get():
        keys = pygame.key.get_pressed()
        if event.type == pygame.QUIT:
            run = False
        if event.type == pygame.KEYDOWN:
            if keys[pygame.K_LEFT]:
                jetx_change = -0.5
            if event.key == pygame.K_SPACE:
                if state == "ready":
                    bulletx = jetx+10
                    fired(jetx,bullety)
                
            if keys[pygame.K_RIGHT]:
                jetx_change = 1
        if event.type == pygame.KEYUP:
            if event.key ==pygame.K_LEFT or event.key == pygame.K_RIGHT:
                jetx_change = 0
        

    jetx += jetx_change
    if jetx <=0:
        jetx = 0
    elif jetx >= 750:
        jetx = 750

    if bullety <= 0:
        state = "ready"
        bullety =  460
    if state == "Fire":
        fired(bulletx,bullety)
        bullety -= bullety_change

    for i in range(no_enemy):
        if planey[i] > 470:
            for j in range(6):
                planey[j] = 2000
            game_over()
            break
        planey[i] += 0.01 
        c = collision(planex[i],planey[i],bulletx,bullety)

    
        if c:
            bullety = 460
            state = "ready"
            score += 1
            print(score)
            planex[i] = random.randint(50,750)
            planey[i] = random.randint(0,100)
        planeplayer(planex[i],planey[i],i)
    
    jetplayer(jetx,jety)
    if score < 5:
        no_enemy = 2
        enemy(2)
    else:
        no_enemy = score+1
        enemy(5)
    
    score_val(tx,ty)
    pygame.display.update()

pygame.quit()    
```
